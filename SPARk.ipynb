```python
# Con esta funcion aplico transformaciones en las columnas 
df2=df.withColumn("consecutivocum",col("consecutivocum").cast(IntegerType()))\
    .withColumn("cantidadcum",col("cantidadcum").cast(IntegerType()))\
    .withColumn("expedientecum",col("expedientecum").cast(FloatType()))\
    .withColumn("cantidadcum",col("cantidadcum").cast(IntegerType()))\
    .withColumn("expedientecum",col("expedientecum").cast(FloatType()))\
    .withColumn("fechaactivo",regexp_extract(col("fechaactivo"),"(\d{4}-\d{2}-\d{2})",1))\

df2.select(col("fechaactivo")).show(2)
df2.select(regexp_replace(col("fechaactivo"),"-|:","."),col("fechaactivo")).show(5,False)  #Funcion para remplazar
df2.select(regexp_extract(col("fechaactivo"),"(\d{4}-\d{2}-\d{2})",1),col("fechaactivo")).show(5,False)  #Funcion para extraer
```

```python
#Filtros especiales
df2.select("producto").withColumn("producto",col("producto").contains("POMADA")).where("producto").show(2)
df2.select(col("expediente").rlike("^2|^8"),col("expediente")).show(10)    #Seleccionar como se hace en sql
```

```python
#Transformar datos de timestram
df2[["fechaactivo"]]
df2.select(col("fechaactivo").cast(TimestampType())).show(3)
df2.select(to_date(col("fechaexpedicion"), "MM/dd/yyyy").cast(TimestampType()),col("fechaexpedicion")).show(3)    #year() funcion para buscar los años
```

```python
#Utilizando funciones where, split
df2.select(col("expediente")).where(expr("expediente like '%8'")).show(5) 
df2.select(split(col("titular")," ")[0],col("titular")).show()

# Complex df
dfcomplex=df2.select(struct("expediente","titular").alias("complex"))
dfcomplex.createOrReplaceTempView("dfcomplex")
display(dfcomplex.show())
```
